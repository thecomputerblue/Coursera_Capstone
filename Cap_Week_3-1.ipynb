## Webscrape Toronto Neighborhoods With BS

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 125)
```

```python
r2 = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
page = bs(r2.content)
postal = page.find("table")
rows3 = postal.find_all("tr")
```

```python
how = rows3[0].find_all("td")
```

```python
how[2].get_text(strip=True)
```

### go through all the rows and take the data tag, then split the postal code from the text into a list

```python
table3 = []

for i in range(len(rows3)):
    sublist = rows3[i].find_all('td')
    for j in range(len(sublist)):
        raw = sublist[j].get_text(strip=True)
        table3.append([(raw[0:3]), (raw[3:])])
    
print(table3)
```

### CREATE THE DATAFRAME FROM THE SCRAPE

```python
t3 = pd.DataFrame(data= table3)
t3.rename(columns= {0: 'Postal_Code', 1: 'Area'}, inplace=True)
t3.head()
```

Split the borough and neighborhoods on the ( character. 

```python
t3['Borough'] = t3['Area'].apply(lambda x: x.split('(')[0])
```

```python
t3['Neighborhood'] = t3['Area'].apply(lambda x: x.split('(')[1] if len(x.split('(')) > 1 else "Not Assigned")
```

drop the raw column

```python
t3 = t3.drop('Area', axis=1)
```

Replace the splitting character and replace the errant )

```python
t3.head(100)
```

```python
t3['Neighborhood'] = t3['Neighborhood'].apply(lambda x: x.replace("/", ", "))
```

```python
t3['Neighborhood'] = t3['Neighborhood'].apply(lambda x: x.replace(")", ""))
```

Drop the unassigned borough rows. 

```python
t3 = t3[t3['Borough'] != "Not assigned"]
```
